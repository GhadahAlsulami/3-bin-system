```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Paths

dataset_dir = r"C:\Users\user\Desktop\capstone\dataset"
output_dir = r"C:\Users\user\Desktop\capstone\split_dataset"

categories = ["Black Box", "Blue Box", "Green Box", "Tuwaiq Cup"]  # Folder names for each class

# Create output directories for YOLO structure
for split in ["train", "val"]:
    for subdir in ["images", "labels"]:
        for category in categories:
            os.makedirs(os.path.join(output_dir, split, subdir, category), exist_ok=True)

# Function to split and copy files
def split_and_copy_files(category):
    category_path = os.path.join(dataset_dir, category)
    all_images = [f for f in os.listdir(category_path) if f.endswith(('.jpg', '.png', '.jpeg'))]

    # Split data into train (80%) and val (20%)
    train, val = train_test_split(all_images, test_size=0.2, random_state=42)

    # Helper to copy files
    def copy_files(file_list, split):
        for file in file_list:
            src = os.path.join(category_path, file)
            dest = os.path.join(output_dir, split, "images", category, file)
            shutil.copy(src, dest)

    # Copy files
    copy_files(train, "train")
    copy_files(val, "val")

# Process each category
for category in categories:
    split_and_copy_files(category)

print("Dataset successfully split into train and validation sets!")
```
